```python
unique_parties=data["Party"].unique()
len(unique_parties)
```

```python
for value in unique_parties:
    print(value, ":", (data["Party"]==value).sum())
```

```python
party_freq_data = []
```

```python
for party in unique_parties:
    frequency = (data["Party"] == party).sum()
    party_data = {"Party": party, "Frequency": frequency}
    party_freq_data.append(party_data)

party_freq = pd.DataFrame(party_freq_data)
```
